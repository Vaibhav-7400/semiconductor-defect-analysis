# Semiconductor Defect Exploratory Analysis

Use this notebook to inspect class balance, feature distributions, and early model behavior.

```python
from pathlib import Path

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data_path = Path('..') / 'data' / 'sample_data.csv'
df = pd.read_csv(data_path)
df.head()
```

```python
df['defect_type'].value_counts().plot(kind='bar', title='Defect Class Balance')
plt.ylabel('wafer count')
plt.tight_layout()
```

```python
numeric_columns = df.select_dtypes(include='number').columns
sns.pairplot(df, vars=numeric_columns[:4], hue='defect_type')
```
